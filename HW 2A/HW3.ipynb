```python
#HW2A.3. Speed of a particle with functional position vector
from sympy import *
t = symbols('t', real = True) #real = True：限定t是实数（而非复数)
r = Matrix([3*cos(2*t), 3*t**2 + t - 3, 3*t**2 + 3*t - 1])
t_v = 2
v = r.diff(t)
v_t=v.subs(t, t_v).norm()
v_t=v_t.evalf()  #evalf()：将符号表达式转换为数值表达式
print(v_t)
```

```python
#HW2A.7. Radial acceleration of a point on a polar curve
import numpy as np
from sympy import *

theta = symbols('theta')

r = 4+ sin(5*theta+4)

# 计算dr/dθ和d²r/dθ²
dr_dtheta = diff(r, theta)
d2r_dtheta2 = diff(dr_dtheta, theta)

# 代入θ=3rad
theta_v = 1
omegaz = 1  # rad/s
alphaz = -2   # rad/s²

# 计算dr/dθ在θ=3处的值
dr_dtheta_val = dr_dtheta.subs(theta, theta_v).evalf()

# 计算d²r/dθ²在θ=3处的值
d2r_dtheta2_val = d2r_dtheta2.subs(theta, theta_v).evalf()

# 计算径向速度r_dot
r_dot = dr_dtheta_val * omegaz

# 计算径向加速度r_ddot
r_ddot = d2r_dtheta2_val * (omegaz**2) + dr_dtheta_val * alphaz

print(f"dr/dθ at θ=3: {dr_dtheta_val}")
print(f"d²r/dθ² at θ=3: {d2r_dtheta2_val}")
print(f"径向速度 r_dot: {r_dot}")
print(f"径向加速度 r_ddot: {r_ddot}")

# 判断运动方向和速率变化
if r_dot > 0:
    direction = "moving away from the origin"
else:
    direction = "moving towards the origin"

if r_ddot * r_dot > 0:
    rate = "at an increasing rate"
else:
    rate = "at a decreasing rate"

print(f"\nConclusion: The particle is {direction} {rate}.")
```

```python
#HW2A.8. Velocity of a point on a polar curve, in Cartesian
import numpy as np
from sympy import *

# 定义r(theta)
r = 4 +sin(4*theta+5)

# 计算dr/dθ
dr_dtheta = diff(r, theta)


theta_v = 1
omegaz = 2

# 计算r在θ=-1处的值
r_val = r.subs(theta, theta_v).evalf()

# 计算dr/dθ在θ=-1处的值
dr_dtheta_val = dr_dtheta.subs(theta, theta_v).evalf()

# 计算径向速度r_dot = dr/dθ * dθ/dt = dr/dθ * ω_z
r_dot = dr_dtheta_val * omegaz

# 极坐标速度分量
v_r = r_dot
v_theta = r_val * omegaz

# 转换为直角坐标分量
theta_rad = float(theta_v)
v_x = v_r * np.cos(theta_rad) - v_theta * np.sin(theta_rad)
v_y = v_r * np.sin(theta_rad) + v_theta * np.cos(theta_rad)

print(f"v_x = {v_x}")
print(f"v_y = {v_y}")
print(f"\n速度矢量: v = {v_x} î + {v_y} ĵ m/s")
```

```python
#HW2A.9. Vector expression derivatives
import numpy as np
from sympy import *

a = Matrix([ 4, -1,  0])
aDot = Matrix([ 0, -5,  0])

amd=aDot.dot(a)/a.norm()
ahd=(aDot*a.norm()-amd*a)/a.norm()**2

print(amd.evalf())
print(ahd.evalf())
```
