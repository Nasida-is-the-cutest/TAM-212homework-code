```python
#HW3
import numpy as np
theta = np.radians(160) # in degrees
r = np.array([2, 1, 0])
u = np.array([np.cos(theta), np.sin(theta), 0])
v = np.array([-np.sin(theta), np.cos(theta), 0])
A=np.array([[u[0],v[0]],[u[1],v[1]]])
b=np.array([r[0],r[1]])
x=np.linalg.solve(A,b)
print(x)  # x[0] is the coordinate in u direction, x[1] is the coordinate in v direction
```

```python
#HW4
import numpy as np

a = np.array([-2,1,0])
b = np.array([a[1],-a[0],0])
c_dot_a = -3
c_dot_b = 10
A=np.array([[a[0],a[1]],
            [b[0],b[1]]])
b=np.array([c_dot_a,c_dot_b])
x=np.linalg.solve(A,b)
print(x)  # x[0] is the coordinate in a direction, x[1] is the coordinate in b direction
```

```python
#HW5
import numpy as np

theta = np.radians(135) # in degrees
rOQ = np.array([-3, -1,  0])
rQP = np.array([ 1, -1,  0])
OP= rOQ + rQP
u = np.array([np.cos(theta), np.sin(theta), 0])
v = np.array([-np.sin(theta), np.cos(theta), 0])
A=np.array([[u[0],v[0]],[u[1],v[1]]])
b=np.array([OP[0],OP[1]])
x=np.linalg.solve(A,b)
print(x)
```

```python
#HW6  e_theta
import numpy as np

Px = -1
Py = 9
v = np.array([-6,  8,  0])

e=np.array([-Py,Px,0])
e_norm = np.linalg.norm(e)
e_unit = e / e_norm
proj= np.dot(v, e_unit)
proj_vector = proj * e_unit
print(proj_vector)
```
