```python
#HW 1B.3
import numpy as np

# 定义已知向量和参数
r = np.array([-4, -5, 0])  # r = 4î - 4ĵ + 0k (m)
omega_z = 4              # ω_z = -3 rad/s

# 计算 r 的垂直向量 r_perp（逆时针旋转90°）
r_perp = np.array([-r[1], r[0], 0])

# 计算速度向量 v = ω_z * r_perp
v = omega_z * r_perp

# 输出结果
print(f"v = {v[0]} î + {v[1]} ĵ + {v[2]} k (m/s)")
```

```python
#HW 1B.5
import numpy as np
a = np.array([2,3,  0])
bLength = 2.000
theta = -0.698 # in radians

# 1. 计算向量a的模长（三维向量，z分量为0，结果等价于二维模长）
a_magnitude = np.linalg.norm(a)

# 2. 计算a方向的单位向量â（三维，z分量保持0）
a_hat = a / a_magnitude

# 3. 计算â的垂直单位向量â⊥（xy平面内逆时针旋转90°，z分量仍为0）
# 三维向量垂直规则：[-a_hat[1], a_hat[0], 0]
a_perp_hat = np.array([a_hat[1], -a_hat[0], 0])

# 4. 计算向量b的分量（核心公式：b = |b|*(cosθ·â + sinθ·â⊥)）
b = bLength * (np.cos(theta) * a_hat + np.sin(theta) * a_perp_hat)

# 5. 输出结果（保留4位小数，更易读）
print(f"向量b的分量：")
print(f"b_x = {b[0]:.4f}")
print(f"b_y = {b[1]:.4f}")
print(f"b_z = {b[2]:.4f}")

# z分量恒为0，符合xy平面运动规律

```

```python
import numpy as np

# 严格使用指定的输入数组
rij = np.array([5.527, 1.409])
ruv = np.array([-5.519, 1.438])

# 1. 构建线性方程组 Ax = b，求解 cosθ 和 sinθ
# 方程组推导：
# rij[0] = cosθ * ruv[0] - sinθ * ruv[1]
# rij[1] = sinθ * ruv[0] + cosθ * ruv[1]
# 整理为：
# ruv[0]*cosθ - ruv[1]*sinθ = rij[0]
# ruv[1]*cosθ + ruv[0]*sinθ = rij[1]
A = np.array([
    [ruv[0], -ruv[1]],  # 第一行：ruv[0]*cosθ - ruv[1]*sinθ
    [ruv[1],  ruv[0]]   # 第二行：ruv[1]*cosθ + ruv[0]*sinθ
])
b = rij  # 等号右侧的 rij 向量

# 2. 解线性方程组，得到 [cosθ, sinθ]
cos_sin_theta = np.linalg.solve(A, b)
cos_theta = cos_sin_theta[0]
sin_theta = cos_sin_theta[1]

# 3. 计算旋转角 θ（弧度），arctan2(sinθ, cosθ) 可正确确定象限
theta = np.arctan2(sin_theta, cos_theta)

# 4. 若 θ 为负，转换到 0~2π 范围（可选，符合常规角度表示）
if theta < 0:
    theta += 2 * np.pi

# 5. 输出结果（保留10位小数，保证精度）
print(f"cosθ = {cos_theta:.10f}")
print(f"sinθ = {sin_theta:.10f}")
print(f"旋转角 θ = {theta:.10f} rad")
```
