```python
#HW4B.2. Tip acceleration for a two-bar linkage
from sympy import *

rOP = Matrix([-1,  3,  0])
rPQ = Matrix([1, 1, 0])
omega1 = Matrix([0, 0, 2])
omega2 = Matrix([ 0,  0, -2])
alpha1 = Matrix([ 0,  0, -3])
alpha2 = Matrix([ 0,  0, -2])


# a_Q = a_P + α × r_PQ + ω × (ω × r_PQ)
ap=alpha1.cross(rOP)+omega1.cross(omega1.cross(rOP))
print(ap)
aq=ap+alpha2.cross(rPQ)+omega2.cross(omega2.cross(rPQ))
print(aq)
```

```python
#HW4B.3. Angular velocity on a two-bar linkage
from sympy import *

rOP = Matrix([2, 1, 0])
rPQ = Matrix([ 0, -2,  0])
vQ = Matrix([-1, -6,  0])

#  v_Q =  v_P + ω × r_PQ
omega1z, omega2z , vpx,vpy= symbols(' omega1z, omega2z ,vpx, vpy')
omega1 = Matrix([0,0,omega1z])
omega2 = Matrix([0,0,omega2z])
vp = Matrix([vpx,vpy,0])
eq1 = Eq(vp, omega1.cross(rOP))
eq2 = Eq(vQ, vp+omega2.cross(rPQ))
solution = solve((eq1,eq2),(omega1z, omega2z , vpx,vpy),dict=True)
print(omega1.subs(solution[0]))
print(omega2.subs(solution[0]))

```

```python
#HW4B.4. Rod acceleration from point acceleration directions
from sympy import Matrix as m
import numpy as np

from sympy import *

l = 5.5
aPMag = 2.0
thetaP = 91.0 # in degrees
aQMag = 3.0
thetaQ = 52.0 # in degrees

thetaP=np.radians(thetaP)
thetaQ=np.radians(thetaQ)

ap=m([aPMag*np.cos(thetaP), aPMag*np.sin(thetaP), 0])
aq=m([-aQMag*np.cos(thetaQ),-aQMag*np.sin(thetaQ),0])
r=m([l,0,0])


alpha,omega=symbols('alpha, omega')
alpha2 = Matrix([0, 0, alpha])
omega2 = Matrix([0, 0, omega])

eq1 = Eq(aq, ap+alpha2.cross(r)+omega2.cross(omega2.cross(r)))
solution = solve((eq1),(alpha, omega),dict=True)

print(alpha2.subs(solution[0]))
print(omega2.subs(solution[0]))
```

```python
#HW4B.5. Four-bar linkage angular velocities
from sympy import *

rAB = Matrix([1, 0, 0])
rBC = Matrix([0, 2, 0])
rAD = Matrix([0, 1, 0])
rDC = Matrix([1, 1, 0])
omega1 = Matrix([0, 0, 3])

omegaz2,omegaz3=symbols('omegaz2, omegaz3')
omega2=Matrix([0,0,omegaz2])
omega3=Matrix([0,0,omegaz3])
vd=omega1.cross(rAD)
vc=vd+omega3.cross(rDC)
vc2=omega2.cross(rBC)
eq1 = Eq(vc, vc2)
solution = solve((eq1),(omegaz2, omegaz3),dict=True)
print(omega2.subs(solution[0]))
print(omega3.subs(solution[0]))
```
