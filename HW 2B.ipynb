```python
#HW2B.2. Velocity of a rotating particle
import numpy as np

x = 0
y = -1
z = -2
omega = np.array([ 1, -2,  2])
r = np.array([x,y,z])
v = np.cross(omega,r) #v=w*r
print(v)
```

```python
#HW2B.4. Particle's velocity on a curve, graphical
from sympy import *
import numpy as np

x = symbols('x')
y = -sin(x)
vx = -5
x_v = 3*np.pi
#vy=dy/dt=dy/dx*(dx/dt)=dy/dx*v
dy=diff(y,x)
x2=dy.subs(x,x_v)
v=vx*x2
print(v)
```

```python
#HW2B.5. Location of a rotating point in the plane
import numpy as np

v = np.array([1, 0, 0])
omegaz = 5
w = np.array([0,0,omegaz])
# r=ai+bj
b=v[0]/-w[2]
a=v[1]/w[2]
print(a)
print(b)

```

```python
#HW2B.6. Particle's acceleration on a curve, graphical
import numpy as np
from sympy import *

x = symbols('x')
y = 2*cos(0.5*x)
vxMag = 4

x_p=np.pi*2 #看图
yp=diff(y,x)
ypp=diff(yp,x)
a=vxMag**2*ypp
ay=a.subs(x,x_p)
print(ay)
```

```python
#HW2B.7. Angular velocity magnitude from speed
import numpy as np

x = 3
y = 0
z = 3
vMag = 5
u = np.array([-3,  1, -5])


length_u=np.linalg.norm(u)
u_hat=u/length_u
r=np.array([x,y,z])
v1=np.cross(u_hat,r) #v=w*r
w=vMag/np.linalg.norm(v1)
print(w)
```

```python
#HW2B.8. Position from polar velocity
#代码需要根据题目数据调整
from scipy import integrate
x=-5
y=3
#v=r*(dtheta/dt)
t=4

r=(x**2+y**2)**0.5
def w(x):
    return -np.e**(x)/r
d=integrate.quad(w,0,t)[0]
theta_initial=np.arctan2(y/r,x/r)
theta_final=theta_initial+d

x2=r*np.cos(theta_final)
y2=r*np.sin(theta_final)
print(x2)
print(y2)


```

```python

import numpy as np

# 1. 初始条件
x0, y0 = 12, 12
t_target = 3

# 2. 初始极坐标
r0 = np.sqrt(x0**2 + y0**2)  # 初始极径
theta0 = np.arctan2(y0, x0)  # 初始角度（π/4）

# 3. 求 r(t)：径向速度 ṙ=1，积分得 r(t) = t + r0
def r(t):
    return t + r0

# 4. 求 θ(t)：角向速度 θ̇ = -2/r(t)，积分得 θ(t) = θ0 - 2*ln((t + r0)/r0)
def theta(t):
    return theta0 - 2 * np.log((t + r0) / r0)

# 5. 计算 t=3 时的极坐标
r3 = r(t_target)
theta3 = theta(t_target)

# 6. 极坐标转直角坐标
x3 = r3 * np.cos(theta3)
y3 = r3 * np.sin(theta3)

# 输出结果
print(f"t={t_target}s 时:")
print(f"r = {r3:.4f} m, θ = {theta3:.4f} rad")
print(f"x = {x3:.4f} m, y = {y3:.4f} m")
```
