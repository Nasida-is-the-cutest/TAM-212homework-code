```python
#HW3A.7. Normal basis vector for a polynomial trajectory
import numpy as np
from sympy import *

t = symbols('t', real = True)
r = Matrix([3*t**2 - 3, -t, 0])
t_v = 1


v = diff(r,t)
et = v.subs(t,t_v)
#print(et)
en = Matrix([-et[1],et[0],0])
length=(en[0]**2+en[1]**2)**0.5
en=en/length
print(en)
```
