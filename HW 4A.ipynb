
方法一：投影法（最直观）
#既然两点之间的距离不变，那么对于刚体上的任意两点，它们各自的速度在两点连线方向上的“投影（分量）”必须完全相等且同#向。否则，这两点就会被拉开或者挤压变形。

方法二： v_Q =  v_P + ω × r_PQ
#* **目测向量减法：** 
  $\vec{v}_Q$ 是向右上方的箭头，$-\vec{v}_P$ （即 $\vec{v}_P$ 的反向）是向左上方的箭头。将这俩相加（即求 $\vec{v}_Q - \vec{v}_P$），得到的结果 $\vec{v}_{Q/P}$ 必然是一个**强烈指向上方、可能略微偏右**的向量。

* **刚体的垂直法则：** 
  在刚体中，一个点相对于另一个点只能做圆周运动。因此，**相对速度的方向，必定永远垂直于这两点的连线！**

```python
#HW4A.6. Center velocity for a two-bar linkage
from sympy import *

rPA = Matrix([ 2, -3,  0])
rQA = Matrix([ 1, -1,  0])
vP = Matrix([-3, -1,  0])
vQ = Matrix([-3, -3,  0])

vAx, vAy, vQx, omega1z, omega2z = symbols('vAx, vAy, vQx, omega1z, omega2z')

# vector unknowns
vA = Matrix([vAx,vAy,0])
omega1 = Matrix([0,0,omega1z])
omega2 = Matrix([0,0,omega2z])

# equations
eq1 = Eq(vA, vP+omega1.cross(rPA))
eq2 = Eq(vA, vQ+omega2.cross(rQA))

# solve
soln = solve((eq1,eq2), (vAx, vAy, omega1z, omega2z), dict=True)
display(soln)

display(vA.subs(soln[0]))
```

```python
from sympy import symbols, Matrix

w_PA = symbols('w_PA')  
w_QA = symbols('w_QA') 
vAx, vAy = symbols('vAx vAy')

rPA = Matrix([0, 2, 0])
rQA = Matrix([-3, -3, 0])
vP = Matrix([-3, 2, 0])
vQ = Matrix([-1, 0, 0])
vA = Matrix([vAx, vAy, 0])

omega_PA = Matrix([0, 0, w_PA])
omega_QA = Matrix([0, 0, w_QA])

eq1 = vA - (vP + omega_PA.cross(rPA))

eq2 = vA - (vQ + omega_QA.cross(rQA))

system = [eq1[0], eq1[1], eq2[0], eq2[1]]

solution = solve(system, (vAx, vAy, w_PA, w_QA))

print(f"点 A 的速度向量为: vA = {solution[vAx]} i + {solution[vAy]} j m/s")
print(f"杆 PA 的角速度: {solution[w_PA]} rad/s")
print(f"杆 QA 的角速度: {solution[w_QA]} rad/s")
```

```python
#HW4A.7. Constrained pin velocity of a rigid body
from sympy import symbols, Matrix, solve
import numpy as np

rPQ = np.array([-2, -1, 0])
vP = np.array([ 2, -1,  0])


#  v_Q =  v_P + ω × r_PQ
t=symbols('t')  # vQ的未知标量
w=symbols('w')  # 角速度的未知标量

vQ=np.array([t,t,0])   
dv=vQ - vP
omega=np.array([0,0,w])
cross_product=np.cross(omega,rPQ)

eq1=dv[0] - cross_product[0]
eq2=dv[1] - cross_product[1]
solution=solve((eq1,eq2),(t,w))
print("杆件的角速度 ω =", solution[w].evalf())


v_Q=(vQ[0]**2 + vQ[1]**2)**0.5
v=v_Q.subs(t, solution[t].evalf())
print("点Q的速度 v_Q =", v)
```

 $\boldsymbol{\vec{a}}_Q = \boldsymbol{\vec{a}}_P + \boxed{\alpha} \times \boldsymbol{\vec{r}}_{PQ} + \boxed{\omega} \times (\boxed{\omega} \times \boldsymbol{\vec{r}}_{PQ})$，  

 $\boldsymbol{\vec{a}}_P$ 是基点$P$的加速度，  

$\boxed{\alpha}$ 为刚体角加速度，  

$\boxed{\omega}$ 为角速度，  

$\boldsymbol{\vec{r}}_{PQ}$ 是从$P$指向$Q$的位置矢量；


结合题目给出的**零角加速度（$\boxed{\alpha}=0$）**条件，切向加速度项 $\boxed{\alpha} \times \boldsymbol{\vec{r}}_{PQ} = 0$，公式简化为 $\boldsymbol{\vec{a}}_Q = \boldsymbol{\vec{a}}_P + \boldsymbol{\vec{a}}_{QP}^n$，其中 $\boldsymbol{\vec{a}}_{QP}^n = \boxed{\omega} \times (\boxed{\omega} \times \boldsymbol{\vec{r}}_{PQ})$ 是$Q$相对$P$的法向（向心）加速度，其物理意义为$Q$绕$P$转动的指向基点的加速度，方向严格沿$PQ$连线、由$Q$指向$P$（与$\boldsymbol{\vec{r}}_{PQ}$反向），结合题图中$P$在$Q$的右上方，因此$\boldsymbol{\vec{a}}_{QP}^n$ 需从$Q$出发沿$PQ$线指向右上方；同时题目给出大小关系 $2\omega^2 r_{PQ} = a_P$，因此法向加速度的大小 $a_{QP}^n = \omega^2 r_{PQ} = \frac{a_P}{2}$，即 $\boldsymbol{\vec{a}}_{QP}^n$ 的长度为 $\boldsymbol{\vec{a}}_P$ 的一半；

画图时需遵循矢量加法的平行四边形法则：第一步，将$P$点的加速度$\boldsymbol{\vec{a}}_P$完整平移到$Q$点（以$Q$为起点，保持与原$\boldsymbol{\vec{a}}_P$完全相同的方向和长度，即水平向右）；第二步，以$Q$为起点，沿$PQ$连线（$Q$指向$P$的方向）画出长度为$\boldsymbol{\vec{a}}_P$一半的法向加速度$\boldsymbol{\vec{a}}_{QP}^n$；第三步，以平移后的$\boldsymbol{\vec{a}}_P$和$\boldsymbol{\vec{a}}_{QP}^n$为邻边作平行四边形，从$Q$出发的对角线即为$Q$点的加速度$\boldsymbol{\vec{a}}_Q$，其方向为水平向右与右上方之间的对角线方向，长度为对角线长度，最终得到题图中粉色箭头所示的正确结果；需注意避免常见错误：不可将法向加速度方向画反（如题图中错误的$\boldsymbol{\vec{a}}_{Q,\text{guess}}$ 方向为左下，违背向心加速度指向基点的要求）、不可错误调整$\boldsymbol{\vec{a}}_{QP}^n$的长度比例、不可额外添加切向加速度项（因$\boxed{\alpha}=0$切向项为零）。
